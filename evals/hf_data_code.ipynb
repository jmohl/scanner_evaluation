```python

# --- Uploading results ----
from pathlib import PurePosixPath

from inspect_ai.log import read_eval_log, write_eval_log

log_file = "core_bench/eval-logs/2026-02-27T20-08-12+00-00_core-bench_ZvU684YYCTbNLk8jJyRdRM.eval"
log = read_eval_log(log_file)
task_name = log.eval.task.split("/")[-1]
filename = PurePosixPath(log_file).name

write_eval_log(
    log,
    location=f"hf://datasets/arcadia-mars-4-0/abc-scout-scanners/{task_name}/eval-logs/{filename}",
)

from pathlib import Path
from inspect_ai._util.file import copy_file

scan_results_base = Path("core_bench/scan-results")
hf_base = "hf://datasets/arcadia-mars-4-0/abc-scout-scanners"

# Derive task name from the parent directory (mirrors eval log upload above)
task_name = scan_results_base.parent.name

for scan_dir in sorted(scan_results_base.glob("scan_id=*")):
    scan_id = scan_dir.name  # e.g. "scan_id=bEiv4f9xi3Kp4VNYo9a9wJ"
    for src_file in sorted(scan_dir.iterdir()):
        dst = f"{hf_base}/{task_name}/scan-results/{scan_id}/{src_file.name}"
        print(f"Uploading {src_file.relative_to(scan_results_base.parent)} → {dst}")
        copy_file(str(src_file), dst)

validation_base = scan_results_base.parent / "validation"
for src_file in sorted(validation_base.iterdir()):
    dst = f"{hf_base}/{task_name}/validation/{src_file.name}"
    print(f"Uploading {src_file.relative_to(validation_base.parent)} → {dst}")
    copy_file(str(src_file), dst)
```
